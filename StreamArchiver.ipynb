# **🚀 StreamArchiver: Ultimate YouTube to Cloud Downloader**
### *Seamlessly download and archive high-quality YouTube content directly to your Google Drive or Local PC.*

---

```python
#@title 🛠️ Setup Environment { display-mode: "form" }
#@markdown Run this cell to install dependencies and optionally mount Google Drive.

from google.colab import drive
from google.colab import files
import os

print("📦 Installing yt-dlp...")
!pip install -q yt-dlp

mount_drive = True #@param {type:"boolean"}
if mount_drive:
    print("📂 Mounting Google Drive...")
    drive.mount('/content/drive')
else:
    print("⏩ Skipping Drive mount. Files will be saved locally in Colab temporary storage.")

print("✅ Setup Complete!")
```

```python
#@title 📥 Download Video/Audio { display-mode: "form" }

video_url = "" # @param {type:"string"}
save_to = "Local PC (Download to Browser)" # @param ["Google Drive", "Local PC (Download to Browser)"]
output_path = "/content/drive/MyDrive/StreamArchiver" # @param {type:"string"}
file_format = "Video + Audio (MP4)" # @param ["Video + Audio (MP4)", "Audio Only (MP3)", "Best Available Quality"]
resolution = "1080p" # @param ["144p", "360p", "480p", "720p", "1080p", "1440p", "2160p (4K)", "Highest Available"]

#@markdown ---
#@markdown *Note: If 'Local PC' is selected, the file will first download to Colab and then trigger a browser download.*

import yt_dlp
import sys
import glob

if not video_url:
    print("❌ Error: Please enter a valid YouTube URL.")
else:
    download_dir = output_path
    if save_to == "Local PC (Download to Browser)":
        download_dir = "/content/downloads"
    
    if not os.path.exists(download_dir):
        os.makedirs(download_dir)

    ydl_opts = {
        'outtmpl': f"{download_dir}/%(title)s.%(ext)s",
        'quiet': False,
        'no_warnings': True,
        'progress_hooks': [lambda d: print(f"📊 Status: {d['_percent_str']} done") if d['status'] == 'downloading' else None],
    }

    if file_format == "Audio Only (MP3)":
        ydl_opts['format'] = 'bestaudio/best'
        ydl_opts['postprocessors'] = [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }]
    else:
        ydl_opts['merge_output_format'] = 'mp4'
        if file_format == "Best Available Quality":
            ydl_opts['format'] = 'bestvideo+bestaudio/best'
        else:
            res_map = {
                "144p": "144", "360p": "360", "480p": "480", "720p": "720", "1080p": "1080", "1440p": "1440", "2160p (4K)": "2160"
            }
            if resolution == "Highest Available":
                ydl_opts['format'] = 'bestvideo+bestaudio/best'
            else:
                res_val = res_map.get(resolution, "1080")
                ydl_opts['format'] = f'bestvideo[height<={res_val}][ext=mp4]+bestaudio[ext=m4a]/best[height<={res_val}]/best'

    print(f"🚀 Starting download for: {video_url}")
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(video_url, download=True)
            filename = ydl.prepare_filename(info)
            # Correct filename extension if merged
            if not file_format == "Audio Only (MP3)":
                filename = os.path.splitext(filename)[0] + ".mp4"
            else:
                filename = os.path.splitext(filename)[0] + ".mp3"

        if save_to == "Local PC (Download to Browser)":
            print(f"📥 Download complete. Transferring to your PC...")
            files.download(filename)
        else:
            print(f"✅ Successfully saved to Google Drive: {filename}")
            
    except Exception as e:
        print(f"❌ An error occurred: {str(e)}")
```
